The covariance is defined as the measure of the joint variability of two random variables.

```python
#Creating an array with all the covariance values
covMat= [C0,C1,C2,C3,C4,C5,C6,C7,C8,C9]

# Computing the inverse of covariance matrices for faster computing
finalCovariances=[]

# Computing the determinants of covariance matrices for faster computing
finalCovDets=[]

for i in range(len(covMat)):
    ranMat=np.zeros(shape=(784,784))
    np.fill_diagonal(ranMat, 0.003)
    finalCovDets.append(np.linalg.slogdet(ranMat+covMat[i]))
    finalCovariances.append(np.linalg.inv(ranMat+covMat[i]))
```

# Why it doesn't perform as good as many other methods on LeCun's web page?

When classifying a given data point, the Naive Baye's classifier estimates the probabilities with which it believes the data points belong to each class label. It selects the class label associated with highest probability. However, beyond the close correlation between the lagest probability and correct label, the probabilities are not correlated with classification confidence.

The Baye's classifier's assumption that features are independent of one another when conditioned upon class labels, is rarely accurate. Features often depend on one another in non-trivial amounts meaning that multiple features often contain similar signals. However, this assumption results in its treatment of features as distinct signals each of which should independently contribute additional confidence to the classifier's prediction.
